```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

Problem 1

Find three simple bivariate functions f1, f2, and f3 and three points p1, p2, p3 such that

p1 is a minimum of f1 p2 is a maximum of f2 p3 is neither a minimum nor maximum of f3, but the gradient of f3 at p3 is the zero vector. Use (a) three-dimensional plots as in three_dimensional_plotting.ipynb and (b) density plots and (c) contour plots as in density_and_contour_plots.ipynb. Indicate clearly the points p1, p2, and p3 in all plots.

Part a)

```
def f(x, y):
    return (2*x)**2 + (2*y)**2 - 4

x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)


X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.scatter(0, 0)
ax.text(0, 3, 0, "P1", color='blue')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.set_title('Part A')
ax.view_init(0, 20)
```

Part b)

```
def f(x, y):
    return np.cos(x)-np.cos(y)

x = np.linspace(-6, 2, 100)
y = np.linspace(-6, 2, 100)


X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.scatter(0, -3.2, 2)
ax.text(1,-3, 2.5, "P2", color='blue')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.set_title('Part B')
ax.view_init(15, 30)
```

Part c)

```
def f(x, y):
    return (x)**2 - y**2

x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)


X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.scatter(0, 0)
ax.text(0,-.5,-10, "P3", color='blue')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.set_title('Part C')
ax.view_init(0, 30)
```
